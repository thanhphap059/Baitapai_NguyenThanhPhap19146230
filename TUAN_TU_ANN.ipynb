```python
#tuan tu mo hinh tao ra ann Sequential (tuần tự)
from keras.models import Sequential
```

```python
#import thư Function API
from keras.layers import Dense, Activation, Dropout 
```

```python
model=Sequential() 
model.add(Dense(512 ,activation='relu', input_shape=(784,))) #784 hang_ 784 tin  hieu vao 512 tin hieu ra
model.add(Dropout(0.2))
model.add(Dense(512 ,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(2,activation='softmax'))
model.summary()
#activation : chọn activation như linear, softmax, relu, tanh, sigmoid
#Dense ( ): Layer này cũng như một layer neural network bình thường, với các tham số sau:
  #units : số chiều output, như số class sau khi train ( chó , mèo, lợn, gà).
  #activation : chọn activation đơn giản với sigmoid thì output có 1 class.
  #use_bias : có sử dụng bias hay không (True or False)
```

```python
from keras.datasets import boston_housing
from keras.models import Sequential
from keras.layers import Dense
```

```python
from tensorflow.keras.optimizers import RMSprop #su dung thu vien toi uu
from keras.callbacks import EarlyStopping #dung tamm thoi nhanh khi xu li du lieu
from sklearn.preprocessing import scale #scale dong ti le
from sklearn.preprocessing import StandardScaler
from sklearn.utils import validation
```

```python
(x_train, y_train),(x_test,y_test)=boston_housing.load_data()
```

```python
x_train_scale=scale(x_train)
```

```python
scaler=StandardScaler().fit(x_train)
x_test_scale=scaler.transform(x_test)
```

```python
#biendich _ chuyen ngu boi co nhieu thong so
model=Sequential()
model.add(Dense(64,kernel_initializer='normal' ,activation='relu', input_shape=(13,)))
model.add(Dense(64 ,activation='relu'))
model.add(Dense(1))
model.summary()
```

```python
from sklearn.utils import validation
model.compile(loss='mse', optimizer=RMSprop(), metrics=['mean_absolute_error']) #matmat,botoiuu
history=model.fit(x_train_scale,y_train,batch_size=128,epochs=500,verbose=1,
                  validation_split=0.2,callbacks=[EarlyStopping(monitor='val_loss',patience = 20)]) #sau nay ve bieu do,solanhoc

#lay du lieu ve do thi, batch size lay 128 du lieu trong 1 lan hoc,epoch  hoc 500 lan,trong qua trinh  hoc danh gia, khi gia tri cal_loss k thay doi trong 20 lan hocthi se dung ngay
```

```python
score= model.evaluate(x_test_scale, y_test,verbose=0)
print('Test loss',score[0])
print('Test accuracy',score[1])

#tinh ra sai so do chinh xac
#so sanh hai gia tri vs nhau
```

```python
y_pred=model.predict(x_test_scale)
print(y_pred.flatten())
print(y_test)
#api
```
